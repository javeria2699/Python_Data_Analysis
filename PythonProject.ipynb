```python
#combine all the filters
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2017) & 
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df1 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
   (df['Year of event'] == 2017) & 
   (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
#remove USA from Event name
```

```python
df1['Event name'] = df1['Event name'].str.split('(').str.get(0)
```

```python
df1.head(5)
```

```python
#cleanup athlete age 
```

```python
df1['athlete_age'] = 2020 - df1['Athlete year of birth'] 
```

```python
#remove h from athlete performance
```

```python
df1['Athlete performance'] = df1['Athlete performance'].str.split(' ').str.get(0) 
```

```python
df1.head(5)
```

```python
# drop columns : Athlete country , Athlete club , Athlete year of birth , Athlete age category
```

```python
df1 = df1.drop(['Athlete club' , 'Athlete country' , 'Athlete age category' , 'Athlete year of birth'], axis = 1)
```

```python
df1.head(3)
```

```python
df1.isna().sum()
```

```python
df1 = df1.dropna()
```

```python
df1.shape
```

```python
#check for dupes
```

```python
df1[df1.duplicated() == True]
```

```python
#reset index
```

```python
df1.reset_index(drop = True)
```

```python
#fix types
```

```python
df1.dtypes
```

```python
df1['athlete_age'] = df1['athlete_age'].astype(int)
```

```python
df1['Athlete average speed'] = df1['Athlete average speed'].astype(float)
```

```python
#rename columns
```

```python
df1 = df1.rename(columns = { 'Year of event': 'year',
    'Event dates': 'event_date',
    'Event name': 'event_name',
    'Event distance/length': 'event_distance',
    'Event number of finishers': 'finishers_count',
    'Athlete performance': 'performance',
    'Athlete gender': 'gender',
    'Athlete average speed': 'average_speed',
    'Athlete ID': 'athlete_id',
    'athlete_age': 'age'              
})
```

```python
df1.head()
```

```python
#reorder columns
```

```python
df3 = df1[['year', 'event_date', 'event_name', 'event_distance', 'finishers_count', 
         'athlete_id', 'gender', 'age', 'average_speed', 'performance']]
```

```python
df3.head()
```

```python
sns.histplot(df3['event_distance'])
```

```python
sns.histplot(df3 , x ='event_distance' , hue = 'gender')
```

```python
sns.distplot(df3[df3['event_distance'] == '50mi']['average_speed'])
```

```python
sns.violinplot(data = df3 , x= 'event_distance' , y = 'average_speed' , hue = 'gender' , split = True , inner='quartile')
```

```python
sns.lmplot(x='age', y='average_speed', data=df3, hue='gender')
```

```python
#difference in speed for 50mi and 50km for male and female
```

```python
df3.groupby(['gender', 'event_distance'])['average_speed'].mean()
```

```python
#what age groups are best in 50mi race
```

```python
df3.query('event_distance == "50mi"').groupby('age')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19')
```
